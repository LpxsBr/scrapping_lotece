```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
chr_opt = Options()
chr_opt.add_argument('--headless')
chr_opt.add_argument('--no-sandbox')
chr_opt.add_argument('--disable-dev-shm-usage')


driver = webdriver.Chrome(options=chr_opt)

driver.get('https://lotece.com.br/resultados/')

dados = (driver.find_element(by=By.CLASS_NAME, value='dataResultado' ).text).split('\n')
driver.quit()
```

```python
data = []
temp = []

for i in dados:
  if str(i)[:str(i).find(' ')].upper() == 'EXTRAÇÃO':
    data.append(temp)
    temp = []
  else:
    temp.append(i)

data.pop(0)

formatedData = []

for y in data:
  formatedData.append(
      {
          "DATA"            : y[0],
          "First"   .upper(): y[3],
          "Second"  .upper(): y[5],
          "Third"   .upper(): y[7],
          "Fourth"  .upper(): y[9],
          "Fifth"   .upper(): y[11],
          "Sixth"   .upper(): y[13],
          "Seventh" .upper(): y[15],
          "Eighth"  .upper(): y[17],
          "Ninth"   .upper(): y[19],
          "Tenth"   .upper(): y[21],
      }
  )
```
